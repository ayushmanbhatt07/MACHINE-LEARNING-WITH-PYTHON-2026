# Working with Dates:

```python
#converting date dataframe to datetime object
```

```python
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

## 1. Extracting year:

```python
date['date_year']=date['date'].dt.year

date.head()
```

## 2. Extracting month:

```python
date['date_month_no.']=date['date'].dt.month

date.head()
```

```python
date['date_month']=date['date'].dt.month_name()

date.head()
```

```python
#day of the month
date['date_day']=date['date'].dt.day

date.head()
```

```python
#day of the week
date['date_dow']=date['date'].dt.dayofweek

date.head()
```

```python
#day of the week name:
date['date_dow_name']=date['date'].dt.day_name()

date.head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)
date.head()
```

## 3.Extracting week of the year:

```python
date['date_week'] = date['date'].dt.isocalendar().week
date.drop(columns=['product_id','city_id','orders']).head()
```

## 4. Extract Quarter:

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

## 5.Extracting semester:

```python
date['semester']=np.where(date['date'].dt.quarter.isin([1,2]),1,2)

date.drop(columns=['product_id','city_id','orders']).head()
```

## 6.Extract time elapsed between two dates(important):

```python
import datetime
today=datetime.datetime.today()
today
```

```python
today-date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed
#this one converts months to be 30days only
np.round((today - date['date']).dt.days / 30, 0)
```

```python
#months passed
#this one gives more accurate vaues
(today.year - date['date'].dt.year) * 12 + (today.month - date['date'].dt.month)
```

# Working on time:

```python

time.info()
```

```python
# Converting to datetime datatype
time['date'] = pd.to_datetime(time['date'])
```

```python
time.info()
```

```python
time['hour'] = time['date'].dt.hour
time['min'] = time['date'].dt.minute
time['sec'] = time['date'].dt.second

time.head()
```

## 7.Extracting time part:

```python
time['time'] = time['date'].dt.time

time.head()
```

### Time difference:

```python
today - time['date']
```

```python
# in seconds
(today - time['date'])/np.timedelta64(1,'s')
```

```python
# in minutes

(today - time['date'])/np.timedelta64(1,'m')
```

```python
# in hours

(today - time['date'])/np.timedelta64(1,'h')
```
